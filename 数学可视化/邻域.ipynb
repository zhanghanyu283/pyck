什么是邻域？
以a为中心的任何开区间称为点a的邻域，记作N(a)。

设δ是任一正数，则开区间(a - δ, a+δ)就是点a的一个邻域，这个邻域称为点a的δ邻域。记作N(a,δ)，即N(a,δ)={x|a-δ < x < a+ δ}。

中心与半径
点a称为这邻域的中心，δ称为这邻域的半径。

去心邻域
点a的δ邻域去掉中心a后，称为点a的，表达方法是在U上标一个小的0。

有时把开区间(a - δ, a)称为a的左δ邻域，把开区间(a, a + δ)称为a的右δ邻域。


```python
import matplotlib.pyplot as plt
import numpy as np

# 设定点 P0 的坐标
x0 = 1
y0 = 1
# 设定邻域半径 δ
delta = 0.5

# 生成网格点
x = np.linspace(x0 - delta * 1.5, x0 + delta * 1.5, 100)
y = np.linspace(y0 - delta * 1.5, y0 + delta * 1.5, 100)
X, Y = np.meshgrid(x, y)

# 计算每个网格点到 P0 的距离
distance = np.sqrt((X - x0) ** 2 + (Y - y0) ** 2)

# 创建图形和坐标轴
fig, ax = plt.subplots()

# 绘制邻域（用颜色填充表示）
ax.contourf(X, Y, distance < delta, cmap='YlOrBr', alpha=0.5)

# 绘制点 P0
ax.plot(x0, y0, 'ro', label=f'$P_0({x0}, {y0})$')

# 设置坐标轴标签和标题
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title(f'$\delta$-neighborhood of $P_0$ with $\delta$ = {delta}')

# 显示图例
ax.legend()

# 显示图形
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# 设置图片清晰度
plt.rcParams['figure.dpi'] = 300

# 设定点 P0 的坐标
x0 = 1
y0 = 1
z0 = 1
# 设定邻域半径 δ
delta = 0.5

# 生成三维网格点
x = np.linspace(x0 - delta * 1.5, x0 + delta * 1.5, 100)
y = np.linspace(y0 - delta * 1.5, y0 + delta * 1.5, 100)
z = np.linspace(z0 - delta * 1.5, z0 + delta * 1.5, 100)
X, Y, Z = np.meshgrid(x, y, z)

# 计算每个网格点到 P0 的距离
distance = np.sqrt((X - x0) ** 2 + (Y - y0) ** 2 + (Z - z0) ** 2)

# 创建图形和三维坐标轴
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 绘制邻域（用颜色填充表示），这里以 z=z0 平面为例
ax.contourf(X[:, :, 50], Y[:, :, 50], distance[:, :, 50] < delta, zdir='z', offset=z0, cmap='Blues', alpha=0.7)

# 绘制邻域边界，这里以 z=z0 平面为例
ax.contour(X[:, :, 50], Y[:, :, 50], distance[:, :, 50], levels=[delta], zdir='z', offset=z0, colors='k', linestyles='dashed')

# 绘制点 P0
ax.scatter(x0, y0, z0, c='r', marker='o', label=f'$P_0({x0}, {y0}, {z0})$')

# 设置坐标轴标签和标题
ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('y', fontsize=12)
ax.set_zlabel('z', fontsize=12)
ax.set_title(f'$\delta$-neighborhood of $P_0$ with $\delta$ = {delta}', fontsize=14)

# 显示图例
ax.legend(fontsize=10)

# 显示图形
plt.show()
    
```

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# 设置图片清晰度
plt.rcParams['figure.dpi'] = 300

# 设定点 P0 的坐标
x0 = 1
y0 = 1
z0 = 1
# 设定邻域半径 δ
delta = 0.5

# 创建三维坐标轴
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 生成球体表面的点
u, v = np.mgrid[0:2 * np.pi:50j, 0:np.pi:25j]
x = x0 + delta * np.cos(u) * np.sin(v)
y = y0 + delta * np.sin(u) * np.sin(v)
z = z0 + delta * np.cos(v)

# 绘制球体表面来表示邻域
ax.plot_surface(x, y, z, color='b', alpha=0.3)

# 绘制点 P0
ax.scatter(x0, y0, z0, c='r', marker='o', label=f'$P_0({x0}, {y0}, {z0})$')

# 设置坐标轴标签和标题
ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('y', fontsize=12)
ax.set_zlabel('z', fontsize=12)
ax.set_title(f'$\delta$-neighborhood of $P_0$ with $\delta$ = {delta}', fontsize=14)

# 显示图例
ax.legend(fontsize=10)

# 显示图形
plt.show()
    
```
